```python
#Calculo de variables probabilisticas
promedio=df['Precipitacion Maxima'].mean()
desviacion_standar=df['Precipitacion Maxima'].std()
alfa=desviacion_standar*((6)**0.5)/np.pi
u=promedio-0.5772*alfa
```

```python
#Cálculo de las Precipitaciones Diarias Máximas Probables para distintas frecuencias
periodos_anuales=[2,5,10,15,25,50,100,200,500]
df_periodo_de_retorno = pd.DataFrame({'Periodo de retorno':periodos_anuales })
#variables calculadas
variable_reducida=[]
precip=[]
prob_ocurrencia=[]
correccion_inter=[]
for i in df_periodo_de_retorno['Periodo de retorno']:
    yt=abs(np.log(np.log(i/(i-1))))
    xt=u+yt*alfa
    fxt=np.e**(-np.e**(-(xt-u)/alfa))
    cxt=xt*1.13
    variable_reducida.append(yt)
    precip.append(xt)
    prob_ocurrencia.append(fxt)
    correccion_inter.append(cxt)
#resultados anadidos en el data frame
df_periodo_de_retorno['Variable reducida (yt)']=variable_reducida
df_periodo_de_retorno['Precipitacion (mm)']=precip
df_periodo_de_retorno['Probabilidad de ocurrencia']=prob_ocurrencia
df_periodo_de_retorno['Correccion del intervalo fijo (mm)']=correccion_inter

df_periodo_de_retorno
```

```python
#Coeficientes para las relaciones a la lluvia de duración 24 horas
tiempo=[24,18,12,8,6,5,4,3,2,1]
coef_24=[1,0.9,0.79,0.64,0.56,0.5,0.44,0.38,0.31,0.25]
#Precipitaciones máximas para diferentes tiempos de duración de lluvias (mm)
df_precipitacion_maxima_td = pd.DataFrame({'Tiempo de duracion (hrs)': tiempo,
                                        'Coeficiente':coef_24})


#Calculo de las precipitaciones por periodo
for i in range(0,df_periodo_de_retorno.shape[0]):
    periodo=df_periodo_de_retorno['Periodo de retorno'][i]
    prec_corr=df_periodo_de_retorno['Correccion del intervalo fijo (mm)'][i]
    pmax_i=[]

    for j in range(0,df_precipitacion_maxima_td.shape[0]):
        coeficiente=df_precipitacion_maxima_td['Coeficiente'][j]
        horas=df_precipitacion_maxima_td['Tiempo de duracion (hrs)'][j]
        pmax=prec_corr*coeficiente
        pmax_i.append(pmax)
    df_precipitacion_maxima_td[f'P. Max.(mm) año {periodo}']=pmax_i

df_precipitacion_maxima_td
    
```

```python
#Intensidades de lluvia a partir de Pd, según Duración de precipitación y Frecuencia de la misma
#Intensidad de la lluvia (mm /hr) según el Periodo de Retorno
cabecera=list(df_precipitacion_maxima_td)
cabecera.remove('Tiempo de duracion (hrs)')
cabecera.remove('Coeficiente')

df_intensidad=pd.DataFrame({'Tiempo de duracion (hrs)': tiempo})
per=1
for i in cabecera:
    df_intensidad[f'Imax (mm/hr) año {per}']=df_precipitacion_maxima_td[i]/df_precipitacion_maxima_td['Tiempo de duracion (hrs)']
    per+=1
df_intensidad
```

```python
#regresion logistica para curva IDF
def curva_idf(x,y):
    lnx=x.apply(np.log)
    lny=y.apply(np.log)
    lnx_lny=lnx*lny
    lnx2=lnx**2
    #sumas
    sum_x=x.sum()
    sum_y=y.sum()
    sum_lnx=lnx.sum()
    sum_lny=lny.sum()
    sum_lnx_lny=lnx_lny.sum()
    sum_lnx2=lnx2.sum()
    num=len(x)
    #constantes
    lnd=((sum_lnx_lny*sum_lnx)-(sum_lnx2*sum_lny))/(sum_lnx**2-sum_lnx2*num)
    d=np.e**lnd
    n=(sum_lny-num*lnd)/sum_lnx
    return d,n
```

```python
#Resumen de aplicación de regresión potencial
df_resumen_regresion=pd.DataFrame({'Periodo de retorno': periodos_anuales})

intensidad_por_periodos=list(df_intensidad)
intensidad_por_periodos.remove('Tiempo de duracion (hrs)')

ctte_d=[]
ctte_n=[]
x=df_intensidad['Tiempo de duracion (hrs)']*60

for i in intensidad_por_periodos:
    y=df_intensidad[i]
    d,n=curva_idf(x,y)
    ctte_d.append(d)
    ctte_n.append(n)
df_resumen_regresion['Coeficiente de Regresion (d)']=ctte_d
df_resumen_regresion['Coeficiente de Regresion (n)']=ctte_n

n=abs(df_resumen_regresion['Coeficiente de Regresion (n)'].mean())

df_resumen_regresion
```

```python
#segunda regresion para hallar los coeficientes k y m
x=df_resumen_regresion['Periodo de retorno']
y=df_resumen_regresion['Coeficiente de Regresion (d)']
K,M=curva_idf(x,y)
```

```python
print(f'Los coeficientes para la ecuacion de intensidad son:\n   k = {K}\n   n = {n}\n   m = {M}')
print('I=(K*T^m)/t^n')
```

```python
fin=60
step=2
duracion_en_minutos=np.arange(1, fin+step, step)
#Tabla de intensidades (mm/min)- Tiempo de duración
df_intensidad_final=pd.DataFrame({'Tiempo de duracion (min)': duracion_en_minutos})
for tr in periodos_anuales:
    intensidad_min=[]
    for t in duracion_en_minutos:
        i=(K*tr**M)/t**n
        intensidad_min.append(i)
    df_intensidad_final[f'I(mm/hr) Año {tr}']=intensidad_min

df_intensidad_final.head(10)
```

```python
intensidades_finales=list(df_intensidad_final)
intensidades_finales.remove('Tiempo de duracion (min)')
for i in intensidades_finales:
    plt.plot(duracion_en_minutos, df_intensidad_final[i], label=i, ls='-')
v = [0, 60, 0, 400]
plt.axis(v)
plt.title('Curvas IDF de la cuenca', color='#00008B')
plt.xlabel('Tiempo de duracion (min)')
plt.ylabel('Intensidad (mm/hr)')
#plt.grid(b=True, which='major', color='#666666', linestyle='--')
plt.legend(loc=4)
plt.show()
```
